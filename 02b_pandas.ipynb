```python
import pandas as pd
import numpy as np

#데이터 프레임의 시각화
# matplotlib의 plot 함수를 데이터프레임에 내장시켜 언제든 그래프를 그릴 수 있게 해줌
data1 = [10,20,30,40,50]
data2 = [1.5,2.5,3.5,4.5,5.5]
idx = np.arange(1,6)
df = pd.DataFrame(data1, index=idx)
df.columns = ['int']
df['float'] = data2
print('시각화용 데이터\n', df)
```

```python
# 데이터프레임.plot.그래프유형
import matplotlib.pyplot as plt

df.plot() # 선 그래프
plt.show()
```

```python
df.plot.bar()  # 막대 그래프
plt.show()
```

```python
df.plot.hist()  # 히스토그램
plt.show()
```

```python
df.plot.box()  # 박스 그래프
plt.show()
```
