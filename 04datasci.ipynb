```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# 미국 질병통제센터에서 수집한 연도별 신생아 성별 출생 수 데이터
# birth.csv 를 이용해서 그래프 작성
```

```python
births = pd.read_csv('c:/Java/data/births.csv')
print(births.head())
print(births.tail())
```

```python
# 피벗 테이블로 집계기능을 사용해서 그래프 생성
# 년도별로 그룹핑해서 출생수로 집계
birth_year = births.pivot_table('births', index='year', columns='gender', aggfunc='sum')
birth_year.plot()
```

```python
# 월별로 그룹핑해서 출생수로 집계
birth_month = births.pivot_table('births', index='month', columns='gender', aggfunc='sum') 
birth_month.plot()
```

```python
# groupby 함수를 이용한 집계 처리
birth_year =births.groupby(['year'])['births'].sum()
print('연도별 출생수 \n', birth_year)
birth_year.plot()
```

```python
# 연도별 성별 출생수
births_M = births[births['gender']=='M']
births_F = births[births['gender']=='F']
births_M_year = births_M.groupby(['year'])['births'].sum()
births_M_year.plot()
births_F_year = births_F.groupby(['year'])['births'].sum()
births_F_year.plot()
```

```python
# 타이타닉 데이터 셋을 이용한 예제
titanic = pd.read_csv('c:/Java/data/train.csv')
print(titanic.head())
# 탑승객 중 성별 생존자 평균 비율
survived = titanic[titanic['Survived']==1]
print('생존자수', len(survived))
# 여성 생존자 수
survivedfemale = survived[survived['Sex']=='female']
print('여성 생존자수', len(survivedfemale))
# print(survivedfemale.head())
# 살아남은 여성 비율 
print('살아남은 여성 비율:', len(survivedfemale)/len(survived))
```

```python
# 탑승객 중 나이별 생존자 평균 비율
age_survived = titanic.pivot_table('Survived', index='Age', aggfunc= 'sum')
# print(age_survived)
c = age_survived['Survived']/len(age_survived)
#print(c)
age_survived['percentage'] = c
print(age_survived.head())
```

```python
# 탑승객 중 요금별 생존자 평균 비율
fare_survived = titanic.pivot_table('Survived', index='Fare', aggfunc= 'sum')
# print(age_survived)
c = fare_survived['Survived']/len(fare_survived)
#print(c)
fare_survived['percentage'] = c
print(fare_survived.head())



```
