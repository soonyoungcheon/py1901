```python
# matplotlib
# 파이썬의 대표적인 시각화 도수
# numpy 기반으로 만들어진 다중 플랫폼 데이터 시각화 라이브러리 
# 하지만, 시간이 지남에 따라 인터페이스와 스타일이 고루해짐
# R의  ggplot 처럼 세련되고 새로운 도구의 출현을 기대함
# 추후에 깔끔하고 현대적인 API로 무장한 seaborn 패키지 탄생

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 간단한 선 그래프 
plt.plot([1,4,9,16,25])
plt.show()
# 지정한 자료는 자동적으로 y축으로 지정
# x축 값이 없으면 자동으로 0,1,2,3,.... 로 지정
# 즉, np.array 객체를 인자로 넘기는 경우,
# y축만 설정하면 x축은 자동으로 감지
```

```python
# 명시적 스타일로 지정하기
# color, linewidth, linestyle, markersize
# marker edge color, marker edge width, marker face color
plt.plot([1,2,3,4,5], [10,20,30,40,50], c='m', lw=3, ls='--',
         marker='d',ms=20, mec='k',mew=5, mfc='r')
plt.show()
```

```python
# 그래프 축 제어하기
# figure : matplotlib에서 축, 그래프, 레이블 등
# 모든 객체를 아우르는 하나의 그릇(container)
# fig = plt.figure()  # plot container
# plt.show()         # 빈화면만 출력

x = np.linspace(0,10,1000)  # 지정한 구간을 구간수로 분할
#print('분할된 구간수', x)
#print('sin 함수 값', np.sin(x))
```

```python
fig = plt.figure()
ax = plt.axes()
ax.plot(x, np.sin(x))
plt.grid()  # 모눈 표시
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes()
ax.plot(x, np.cos(x))
plt.grid()  # 모눈 표시
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes()
ax.plot(x, np.tan(x))
plt.grid()  # 모눈 표시
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes()
ax.plot(x, np.tan(x), 'r:',
        x, np.cos(x),'b--',
        x,np.tan(x/2),'gp-.')
plt.grid()  # 모눈 표시
plt.ylim(5,-5)
plt.xlim(0,10)
plt.show()
```

```python
# 그래프 색상 지정 방식
fig = plt.figure()
ax = plt.axes()
ax.plot(x,np.sin(x), c='red') # 색상명
ax.plot(x,np.sin(x), c='b')  # 짧은 색상명
ax.plot(x,np.sin(x), c=(1.0,0.2,0.3))  # RGB tuple (0,1)
ax.grid()
plt.axhline(y=0, c='b', linewidth=1)
```

```python
# 그래프 선 모양 지정 방식
fig = plt.figure()
ax = plt.axes()
ax.plot(x, x + 2, linestyle='solid') # -
ax.plot(x, x + 4, linestyle='dashed') # --
ax.plot(x, x + 6, linestyle='dotted') # .
ax.plot(x, x + 8, linestyle='dashdot') # -.
plt.show()
```

```python
# 그래프 축, 라벨, 타이틀 지정
fig = plt.figure()
ax = plt.axes()
ax.plot(x, np.sin(x),'r', label='sin(x)')
ax.plot(x, np.cos(x),'b', label='cos(x)')
plt.legend() # 범례
plt.title('sin/cos function curve')  # 제목
plt.xlabel('x value')          # x축 제목      
plt.ylabel('y value')    # y축 제목 
plt.show()
```

```python
# 그래프 축, 라벨, 타이틀 통합 지정 - ax.set
fig = plt.figure()
ax = plt.axes()
ax.plot(x, np.sin(x),'r', label='sin(x)')
ax.plot(x, np.cos(x),'b', label='cos(x)')
plt.legend() # 범례
ax.set(xlim=(0,10),ylim=(-1.1,1.1), xlabel='x value'
       ,ylabel='y value', title='sin/cos function curve')
ax.grid()
plt.show()
```
